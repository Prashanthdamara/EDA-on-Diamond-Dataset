## Exploratory Data Analysis on Diamond Dataset

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv('diamonds.csv')
```

```python
df.head()
```

```python
#check info
df.info()
```

```python
#describe dataset
df.describe()
```

```python
#check null values
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),cbar=False,yticklabels=False)
```

#### There is no null values

```python
df.head()
```

```python
#plot cut variable using barplot
sns.barplot(x=df.cut.value_counts().index, y=df.cut.value_counts())
```

#### There are more ideal cut diamonds in this dataset

```python
#plot color variable using barplot
sns.barplot(x=df.color.value_counts().index, y=df.color.value_counts())
```

#### There are more G type color diamonds in this dataset

```python
#plot clarity variable using barplot
sns.barplot(x=df.clarity.value_counts().index, y=df.clarity.value_counts())
```

#### There are more SI1 clarity type diamonds in this dataset

```python
df.head()
```

```python
#plot distribution of carat
sns.distplot(df.carat)
```

#### This carat distribution is not normal and it is multi-modal

```python
#plot carat varaible using boxplot and check null values
plt.boxplot(df.carat)
plt.show()
```

#### This carat feature has outliers and the median is around 0.5, the most of the data is below 1 and it is right skewed

```python
#plot distribution of price
sns.distplot(df.price)
```

#### The price distribution graph is right skewed and it is not normal 

```python
#plot price variable using boxplot and check outliers
plt.boxplot(df.price)
plt.show()
```

#### The price has outliers, the distribution is right skewed, the median price is around 2500 and the most of the data is below around 5000

```python
df.head()
```

```python
#check relationship between carat and price
sns.lmplot(x = 'carat',y='price',data = df)
```

#### it has positive relationship,it is linear and it has outliers 

```python
#check relationship of cut and price
plt.figure(figsize=(12,5))
sns.lineplot(x = 'cut',y='price',data = df)
```

#### The price for premium cut is high

```python
#check relationship of cut and price
plt.figure(figsize=(12,5))
sns.lineplot(x = 'color',y='price',data = df)
```

#### The price for J color is high

```python
#check relationship of clarity and price
plt.figure(figsize=(12,5))
sns.lineplot(x = 'clarity',y='price',data = df)
```

#### The SI2 type clarity has more price and WS1 has low price

```python
#check relationship of price and color with cut
plt.figure(figsize=(12,5))
sns.lineplot(x = 'color',y='price',data = df,hue = 'cut')
```

#### The price for 'J' color with premium cut is high  and 'D' color with ideal cut is low price

```python
#check relationship of price and color with cut
plt.figure(figsize=(12,5))
sns.lineplot(x = 'color',y='price',data = df,hue = 'clarity')
```

#### The price for D color with 'IF' clarity is high and price for E color with 'ws1' is low

```python
#check relationship of carat and price with color
plt.figure(figsize=(12,5))
sns.scatterplot(x = 'carat',y='price',data = df,hue = 'color')
```

```python
#check relationship of carat with x,y,z variables
plt.figure(figsize=((12,5)))
plt.subplot(1,3,1)
sns.scatterplot(x = 'x',y = 'carat',data = df)
plt.subplot(1,3,2)
sns.scatterplot(x = 'y',y = 'carat',data = df)
plt.subplot(1,3,3)
sns.scatterplot(x = 'z',y = 'carat',data = df)
```

#### This dimensions x,y,z also have positive relationship with carat but x is more influencing carat or viceversa

```python
#check relationship of price with x,y,z variables
plt.figure(figsize=((12,5)))
plt.subplot(1,3,1)
sns.scatterplot(x = 'x',y = 'price',data = df)
plt.subplot(1,3,2)
sns.scatterplot(x = 'y',y = 'price',data = df)
plt.subplot(1,3,3)
sns.scatterplot(x = 'z',y = 'price',data = df)
```

#### This dimensions x,y,z also have positive relationship with price but x is more influencing price or viceversa
#### from above fig carat v/s x,y,z also have same relationship here because as carat refer to weight of diamond so carat increase x (refers length(mm)) increases which increases the price

```python
df.head()
```

```python
#plot heatmap
plt.figure(figsize=(12,5))
sns.heatmap(df.corr(),annot=True)
```

- x,y,z have strong postive relationship with carat and same for price 
- price and carat have strong positive relationship
